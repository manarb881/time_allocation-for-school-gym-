```python
df.rename(columns={'Unnamed: 1':'Sunday ','Unnamed: 2':'Tuesday','14-15, 15-16, 16-17':'Thursday','Lacheb mohamed':'full name'}, inplace=True)
df
```

```python
time_slots = {
    'Sunday': { '11-12': [], '12-13': [], '13-14': [], '14-15': [], '15-16': [], '16-17': [], '17-18': [], '18-19': [] },
    'Tuesday': { '11-12': [], '12-13': [], '13-14': [], '14-15': [], '15-16': [], '16-17': [], '17-18': [], '18-19': [] },
    'Thursday': { '11-12': [], '12-13': [], '13-14': [], '14-15': [], '15-16': [], '16-17': [], '17-18': [], '18-19': [] }
}
type(time_slots )
```

### split the time slots on the three columns 

```python
df.columns = df.columns.str.strip()
```

```python
# Replace NaN values with '0'
df[['Sunday','Tuesday','Thursday']] = df[['Sunday', 'Tuesday', 'Thursday']].fillna('0')

df
```

```python
df['num-available-slots'] = df[['Sunday', 'Tuesday', 'Thursday']].apply(
    lambda row: sum(len(str(day).split(',')) for day in row if day != '0'), axis=1
)
df
```

```python
df=df.sort_values(by='num-available-slots')
df
```

#### delete those with zero hour slots 

```python
df=df.drop(index=df[df['num-available-slots']==0].index)
df = df.reset_index(drop=True)
df        
```

```python
df.tail()
```

#### creating new table to store final schedule hours in it 

```python
new_df=df
```

```python
# Initialize a dictionary to keep track of the number of participants assigned to each time slot
slot_capacity_sunday = {
    '11-12': 0,
    '12-13': 0,
    '13-14': 0,
    '14-15': 0,
    '15-16': 0,
    '16-17': 0,
    '17-18': 0,
    '18-19': 0,
    # Add other time slots as necessary
}
slot_capacity_thursday = {
    '11-12': 0,
    '12-13': 0,
    '13-14': 0,
    '14-15': 0,
    '15-16': 0,
    '16-17': 0,
    '17-18': 0,
    '18-19': 0,
    # Add other time slots as necessary
}
slot_capacity_tuesday = {
    '11-12': 0,
    '12-13': 0,
    '13-14': 0,
    '14-15': 0,
    '15-16': 0,
    '16-17': 0,
    '17-18': 0,
    '18-19': 0,
    # Add other time slots as necessary
}
capacity={'Sunday':slot_capacity_sunday,'Tuesday':slot_capacity_tuesday,'Thursday':slot_capacity_thursday} 

# Loop through each single-slot participant
for index, row in df.iterrows():
    # Check available slots for Sunday, Tuesday, and Thursday
    for day in ['Sunday', 'Tuesday', 'Thursday']:
        if row[day] != '0': 
            preferred_slots = str(row[day]).split(', ') 
        
            for slot in preferred_slots:
                if slot in capacity[day] and capacity[day][slot] < 20:
                    # Assign the participant to the slot
                    capacity[day][slot] += 1
                    time_slots[day][slot].append(df.at[index, 'full name']) 
                    new_df.at[index, day] = slot  # Update the DataFrame
                    break  # Move to the next participant after assigning
                
```
